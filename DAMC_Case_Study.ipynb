# Data Cleaning

Data cleaning on case study NYC AirBnB by Annisa Gita Rahayu

```python
import pandas as pd
airbnb = pd.read_csv('NYC AirBnB Dummy Data - DAMC_C.csv', sep = ';', encoding = "unicode_escape")
airbnb.head()
```

Change data type of price and remove the ' dollars'

```python
airbnb.dtypes
```

```python
airbnb['price'] = airbnb['price'].str.strip(' dollars')
airbnb['price'].sum()
```

```python
airbnb['price'] = airbnb['price'].astype('int')
airbnb['price'].dtypes
```

Check the empty data and drop it

```python
airbnb.info()
```

```python
airbnb.isna().sum()
```

```python
airbnb[airbnb.isnull().any(axis=1)]
```

```python
airbnb = airbnb.copy()
airbnb.dropna(axis=0, inplace=True)
airbnb
```

```python
airbnb.info()
```

Check duplicate data

```python
airbnb[airbnb.duplicated()]
```

Datetime

```python
airbnb.head()
```

```python
airbnb['last_review'] ='20' + airbnb['last_review'].astype(str) #menambahkan '20' di awal kata
airbnb['last_review'] = airbnb['last_review'].str.split('/').apply(lambda x: '/'.join(x[::-1])) #membalik urutan
airbnb['last_review'] = airbnb['last_review'].str[2:] #menghapus 2 karakter pertama yaitu '20'
```

```python
airbnb
```

```python
import datetime as dt
airbnb['last_review'] = pd.to_datetime(airbnb['last_review']).dt.date
airbnb.dtypes
```

```python
airbnb.head()
```

```python
today_date = dt.date.today()
airbnb[airbnb['last_review'] > today_date]
```

```python
airbnb.loc[airbnb['last_review'] > today_date, 'last_review'] = today_date
airbnb[airbnb['last_review'] > today_date]
```

```python
airbnb
```

Splitting column

```python
airbnb['neighborhood_full'].str.split(',', expand=True)
```

```python
airbnb[['region','sub_region','none']] = airbnb['neighborhood_full'].str.split(',', expand=True)
```

```python
airbnb.head()
```

Delete column 'none' using **.drop('column')** method

```python
airbnb = airbnb.drop('none', axis =1)
airbnb.head()
```

Correcting the strings

```python
airbnb['room_type'] = airbnb['room_type'].str.title()
```

```python
airbnb.head()
```
